# **Data Preprocessing**

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

### **1. Menghapus Kolom yang Akan Dijadikan Objek Prediksi**

```python
x = df.drop('Attrition', axis=1)
y = df['Attrition']
```

### **2. Melakukan Label Encoding**

Label Encoding adalah teknik untuk mengubah data kategorikal (teks/string) menjadi angka (integer), supaya bisa diproses oleh algoritma machine learning.

```python
cat_cols = x.select_dtypes(include=['object']).columns

le = LabelEncoder()
for col in cat_cols:
    x[col] = le.fit_transform(x[col])
```

### **3. Melakukan Train-Test Split**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### **4. Melakukan Normalisasi Data**

```python
scaller = StandardScaler()
x_train = scaller.fit_transform(x_train)
x_test = scaller.transform(x_test)
```

# **Modelling**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

### **1. Logistic Regression**

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)

print(classification_report(y_test, y_pred))
```

### **2. Random Forest Classification**

```python
rfc = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=42, class_weight='balanced')
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)

print(classification_report(y_test, y_pred))
```

### **3. Naive Bayes Classification**

```python
nvb = GaussianNB()
nvb.fit(x_train, y_train)
y_pred = nvb.predict(x_test)

print(classification_report(y_test, y_pred))
```

### **4. Perbandingan Ketiga Model**

```python
y_pred_lr = lr.predict(x_test)
y_pred_rfc = rfc.predict(x_test)
y_pred_nvb = nvb.predict(x_test)

tabel_akurasi_model = pd.DataFrame({
    "Logistic Regression": [accuracy_score(y_test, y_pred_lr)],
    "Random Forest": [accuracy_score(y_test, y_pred_rfc)],
    "Naive Bayes": [accuracy_score(y_test, y_pred_nvb)]
}, index=["Akurasi"])
tabel_akurasi_model

```

# **Evaluation**

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### **1. Visualisasi Hasil Dalam Bentuk Confusion Matrix**

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Prediksi')
plt.ylabel('Aktual')
plt.title('Confusion Matrix')
plt.show()

# Classification Report
print(classification_report(y_test, y_pred_rfc))
```

### **2. ROC Curve**

ROC (Receiver Operating Characteristic) adalah kurva yang menggambarkan performa model klasifikasi di berbagai threshold keputusan.

Kurva ini memplot True Positive Rate (TPR) terhadap False Positive Rate (FPR) pada berbagai threshold. Semakin mendekati kurva ke kiri atas, semakin baik performa model.

```python
plt.figure(figsize=(8, 6))

# Logistic Regression
y_prob = lr.predict_proba(x_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label=f'Logistic Regression (AUC = {roc_auc_score(y_test, y_prob):.2f})')

# Random Forest
y_prob = rfc.predict_proba(x_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label=f'Random Forest (AUC = {roc_auc_score(y_test, y_prob):.2f})')

# Naive Bayes
y_prob = nvb.predict_proba(x_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label=f'Naive Bayes (AUC = {roc_auc_score(y_test, y_prob):.2f})')

# Garis baseline
plt.plot([0, 1], [0, 1], 'k--', label='Random')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

### **3. Feature Importance**

Feature importance adalah metrik yang digunakan untuk mengukur seberapa penting setiap fitur dalam model machine learning. Semakin tinggi nilai feature importance, semakin penting fitur tersebut dalam model.

```python
feature_importance = pd.DataFrame({'Fitur': x.columns,'Kepentingan' : np.abs(lr.coef_[0])}).sort_values(by='Kepentingan', ascending=False)
feature_importance.head(10)
```

### **4. Top 10 Fitur Paling Penting**

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Kepentingan', y='Fitur', data=feature_importance.head(10))
plt.title("Top 10 Fitur Paling Penting")
plt.show()
```
